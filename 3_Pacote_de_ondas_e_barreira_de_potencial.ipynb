## Potencial e Fun√ß√£o de Onda

Dessa vez o potencial que nossa part√≠cula est√° submetida √© nulo! Ou seja,

$$
V(x) = 0
$$

Dessa forma, a equa√ß√£o de Schrodinger fica mais simples:

$$
-\frac{ƒß^2}{2m} \frac{d^2ùõ∑(x)}{dx^2} = Eùõ∑(x)
$$

Podemos introduzir uma constante para simplificar:

$$
k = \sqrt{\frac{2mE}{ƒß}}
$$

Reorganizando:

$$
\frac{d^2ùõ∑(x)}{dx^2} + k^2ùõ∑(x) = 0
$$

Essa equa√ß√£o √© a mesma que encontramos no Po√ßo de Potencial Infinito, mas agora sem as condi√ß√µes de contorno! Veremos que essa pequena mudan√ßa ir√° alterar significativamente os resultados.

---

## Solu√ß√£o

Novamente, temos a solu√ß√£o geral dada por:

$$
ùõ∑(x) = Ae^{ikx} + Be^{-ikx}
$$

ou

$$
ùõ∑(x) = e^{ikx}
$$

*A primeira √© a forma mais completa, considerando ondas se propagando para direita e esquerda, "indo e vindo". √â mais usada quando queremos descrever fenom√™nos de reflex√£o, espalhamento ou simetria. A segunda √© um caso particular em que B = 0 e A = 1, ou seja, temos apenas uma onda se propagando para direita. Nesse caso, temos um estado de momento definido, nos permitindo montar pacotes de ondas com A(k). N√£o se preocupe, essa "montagem" ser√° explicada em breve!*

N√£o temos condi√ß√µes de contorno para definirmos as constantes A e B! A densidade de probabilidade √© constante para todo x, **a part√≠cula tem a mesma probabilidade de estar em qualquer lugar, ou seja, nossa solu√ß√£o n√£o √© normaliz√°vel!**
Se recorremos ao princ√≠pio da Incerteza, ter√≠amos uma situa√ß√£o onde a incerteza na posi√ß√£o x tenderia ao **infinito**, ou seja, a incerteza no momento p tenderia a 0, **ter√≠amos certeza do momento da part√≠cula!**

Para corrigir esse problema, ou seja, torn√°-lo fisicamente aceit√°vel, montamos uma combina√ß√£o de ondas com diferentes k (n√∫meros de onda) atrav√©s da transformada inversa de Fourier! Dessa forma, geramos pacotes de ondas localizados no espa√ßo, permitindo incertezas finitas tanto na posi√ß√£o quanto no momento.

$$
Œ®(x) = ‚à´ A(k)e^{ikx} dx
$$

Onde A(k) √© a amplitude da contribui√ß√£o de cada momento!

Curiosamente, a energia de cada part√≠cula se assemelha bastante a **Energia Cin√©tica Cl√°ssica!** Basta substituir o caso particular da solu√ß√£o gera (propaga√ß√£o √∫nica para direita) na equa√ß√£o de Schrodinger:

$$
-\frac{\hbar^2}{2m} \frac{d^2\phi(x)}{dx^2} = E\phi(x)
$$

Solu√ß√£o:

$$
\phi_k(x) = e^{ikx} \quad \text{com} \quad k = \frac{p}{\hbar}
$$

Substituindo:

$$
\frac{d^2}{dx^2} e^{ikx} = -k^2 e^{ikx}
$$

$$
\Rightarrow -\frac{\hbar^2}{2m} (-k^2)\cdot e^{ikx} = E \cdot e^{ikx}
$$

$$
\Rightarrow \frac{\hbar^2 k^2}{2m} = E
$$

Pela rela√ß√£o de De Broglie:

$$
p = ƒßk
$$

$$
‚áí E = \frac{p^2}{2m}
$$

Sendo essa a energia de cada part√≠cula livre em fun√ß√£o do momento!

## Intera√ß√£o com uma barreira de potencial - Caso particular

Agora que sabemos como nosso pacote de onda se comporta segundo a Equa√ß√£o de Schrodinger, conseguimos experimentar o que aconteceria caso ele colidisse com uma barreira de potencial.

De forma intuitiva (n√£o vamos tratar muito da matem√°tica do problema por enquanto), a fun√ß√£o de onda precisa existir em todos os locais devido as condi√ß√µes de continuidade da Equa√ß√£o de Schrodinger, ou seja, **ela pode existir atravessando a barreira, mesmo com energia menor**.

O estudo desse fen√¥meno nos revela duas constantes que ser√£o importantes na simula√ß√£o, R e T, Coeficiente de Reflex√£o e Coeficiente de Transmiss√£o, respectivamente. Sua soma deve resultar em 1 (100%), pois a fun√ß√£o n√£o pode simplesmente sumir! Dessa forma, pela continuidade, **sempre vamos ter uma parcela de cada um, nunca apenas transmiss√£o ou reflex√£o, como esperado na mec√¢nica cl√°ssica!**

*Apenas uma observa√ß√£o: Como ainda estamos no densenvolvimento do c√≥digo, pode ser que a soma de R e T nem sempre seja exatamente 1. Fisicamente, isso n√£o √© aceit√°vel, mas aqui poder√° ocorrer pelo fato de que a simula√ß√£o est√° em processo de aprimoramento constante. Pretendemos ajustar isso no futuro!*

## Aplica√ß√µes

Uma das aplica√ß√µes mais interessantes dos pacotes de ondas livres s√£o os aceleradores de part√≠culas e a Engenharia de part√≠culas. Os pacotes de onda nos ajudam a modelar o movimento, espalhamento e interfer√™ncia das part√≠culas.
J√° para o caso de uma barreira de potencial, temos a **fus√£o nuclear no Sol**. Para que o Sol "continue funcionando", os pr√≥tons em seu centro precisam se unir, mas a energia de repuls√£o √© extremamente alta! Classicamente, eles nunca se uniriam, mas qu√¢nticamente vimos que √© poss√≠vel! A continuidade matem√°tica com as condi√ß√µes de contorno nos mostraram que eles podem tunelar por essa barreira, **permitindo a fus√£o e continua√ß√£o do processo!**

Para finalizar, temos uma legenda explicando o que √© cada par√¢metro e as cores da anima√ß√£o:

```python
!pip install -q ipywidgets

# Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from scipy.sparse import diags
from scipy.sparse.linalg import splu
from IPython.display import HTML, display
import ipywidgets as widgets

# Constantes para simplificar
hbar = 1.0
m    = 1.0

# Defini√ß√£o da fun√ß√£o do pacote de onda
def simulate_wavepacket(a=5.0, V0=2.0, k0=2.0, sigma=2.0,
                        dt=0.02, T_max=100.0, Nx=250, L=250.0,
                        absorb_width=50.0, absorb_strength=1e-2):

# Grade espacial
    x  = np.linspace(0, L, Nx)
    dx = x[1] - x[0]
    center = L/2
    steps = int(T_max / dt)

# Potencial: barreira real + absor√ß√£o imagin√°ria
    V = np.zeros(Nx, dtype=complex)
    mask_bar = np.abs(x - center) < (a/2)
    V[mask_bar] = V0
    mask_abs = np.zeros(Nx)
    left  = x < absorb_width
    right = x > (L - absorb_width)
    mask_abs[left]  = ((absorb_width - x[left])  / absorb_width)**2
    mask_abs[right] = ((x[right] - (L-absorb_width)) / absorb_width)**2
    V += 1j * absorb_strength * mask_abs

# Pacote inicial normalizado
    x0 = L/4
    env   = np.exp(-((x - x0)**2)/(2*sigma**2))
    phase = np.exp(1j * k0 * x)
    psi0  = env * phase
    psi0 /= np.sqrt(np.sum(np.abs(psi0)**2) * dx)

# Energia do pacote
    E_packet = hbar**2 * k0**2 / (2*m)
    print(f"E_pac = {E_packet:.3f}, V0 = {V0:.3f}")

# Matrizes CN
    coeff    = hbar**2 / (m * dx**2)
    diag_A   = 1 + 1j*dt/(2*hbar)*(coeff + V)
    diag_B   = 1 - 1j*dt/(2*hbar)*(coeff + V)
    off_val  = -1j*dt/(4*hbar)*coeff
    off_arr  = np.full(Nx-1, off_val, dtype=complex)

    A = diags([off_arr, diag_A, off_arr],
              offsets=[-1, 0, 1],
              shape=(Nx, Nx),
              format='csc')
    B = diags([-off_arr, diag_B, -off_arr],
              offsets=[-1, 0, 1],
              shape=(Nx, Nx),
              format='csc')
    A_fac = splu(A)

# Prepara√ß√£o da barreira
    psi = psi0.copy()
    fig, ax = plt.subplots(figsize=(8,4))
    ax.set_xlim(0, L)
    ax.set_ylim(-1.1, 1.1)
    ax.set_xlabel('x')
    ax.set_ylabel(r'$\Re(\Psi)$ & $|\Psi|^2$')
    V_plot = V.real/np.max(V.real)
    ax.plot(x, V_plot, 'k--', label='Barreira')
    raw_line , = ax.plot([], [], label=r'Re[$\Psi$]')
    prob_line, = ax.plot([], [], label=r'$|\Psi|^2$')
    info_text = ax.text(0.02, 0.95, '', transform=ax.transAxes, va='top')
    ax.legend(loc='upper right')

    def init():
        raw_line.set_data([], [])
        prob_line.set_data([], [])
        info_text.set_text('')
        return raw_line, prob_line, info_text

# Amostragem de no m√°ximo 500 quadros para evitar sobrecarga
    max_frames = 500
    frame_idx  = np.linspace(0, steps-1, min(steps, max_frames), dtype=int)

# Defini√ß√£o da fun√ß√£o da anima√ß√£o
    def update(i):
        nonlocal psi

# Evolu√ß√£o nos frames
        n_steps = frame_idx[i] - (frame_idx[i-1] if i>0 else -1)
        for _ in range(n_steps):
            psi = A_fac.solve(B.dot(psi))

        t = frame_idx[i] * dt
        raw_line.set_data(x, np.real(psi))
        prob_line.set_data(x, np.abs(psi)**2)
        info_text.set_text(f"t = {t:.2f}")

        if i == len(frame_idx)-1:
            left_area  = x < (center - a/2)
            right_area = x > (center + a/2)
            R = np.sum(np.abs(psi[left_area])**2) * dx
            T = np.sum(np.abs(psi[right_area])**2) * dx
            info_text.set_text(
                f"t={t:.1f} | E={E_packet:.2f}, V0={V0:.2f}, R={R:.3f}, T={T:.3f}"
            )

        return raw_line, prob_line, info_text

    ani = animation.FuncAnimation(
        fig, update, frames=len(frame_idx),
        init_func=init, blit=True, interval=30, repeat=False
    )
    plt.close(fig)
    display(HTML(ani.to_jshtml()))

# Interatividade
widgets.interact_manual(
    simulate_wavepacket,
    a=widgets.FloatSlider(value=5.0, min=1.0, max=5.0, step=1.0,   description='Largura da barreira'),
    V0=widgets.FloatSlider(value=2.0,  min=1.0, max=5.0,   step=1.0, description='Potencial da barreira'),
    k0=widgets.FloatSlider(value=2.0,  min=1.0, max=3.0,  step=1.0, description='N√∫mero de onda'),
    sigma=widgets.FloatSlider(value=2.0, min=1.0, max=3.0,  step=1.0, description='Dispers√£o espacial'),
    dt=widgets.FloatLogSlider(value=0.02, base=10, min=-3, max=-1, step=0.1, description='Intervalo de tempo por frame na anima√ß√£o'),
    T_max=widgets.FloatSlider(value=100.0, min=10.0, max=200.0, step=5.0, description='Tempo m√°ximo da anima√ß√£o'),
    Nx=widgets.IntSlider(value=250, min=200, max=400, step=50, description='Suaviza√ß√£o da anima√ß√£o'),
    L=widgets.FloatSlider(value=250.0, min=100.0, max=300.0, step=50.0, description='Largura do eixo x'),
    absorb_width=widgets.FloatSlider(value=50.0,min=10.0, max=100.0, step=5.0, description='Largura de absor√ß√£o do pacote'),
    absorb_strength=widgets.FloatLogSlider(value=1e-2, base=10, min=-4, max=-1, step=0.1, description='For√ßa de absor√ß√£o do pacote'),
)
```
